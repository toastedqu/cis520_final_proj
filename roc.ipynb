```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from sklearn.metrics import roc_curve
```

## Step 1

```python
y = pd.read_csv("test_ImagesAndLabels.csv")[:20000]["label"].to_numpy()
```

```python
prob_log = pickle.load(open("lo_5_2000_prediction_proba.sav", 'rb'))
prob_knn = pickle.load(open("neigh_5_2000_prediction_proba.sav", 'rb'))
prob_svm = pickle.load(open("svc_5_2000_prediction_proba.sav", 'rb'))
prob_xgb = pickle.load(open("xgb_5_2000_prediction_proba.sav", 'rb'))
prob_cnn = np.load("v1_step1_scores.npy")
prob_res = np.load("y_score_res_1.npy")
prob_vit = np.load("y_score_vit_1.npy")
```

```python
fpr_log, tpr_log, _ = roc_curve(y, prob_log[:,1])
fpr_knn, tpr_knn, _ = roc_curve(y, prob_knn[:,1])
fpr_svm, tpr_svm, _ = roc_curve(y, prob_svm)
fpr_xgb, tpr_xgb, _ = roc_curve(y, prob_xgb[:,1])
fpr_cnn, tpr_cnn, _ = roc_curve(y, prob_cnn)
fpr_res, tpr_res, _ = roc_curve(y, prob_res[:,1])
fpr_vit, tpr_vit, _ = roc_curve(y, prob_vit[:,1])
```

```python
plt.plot(fpr_log, tpr_log, label="logreg")
plt.plot(fpr_knn, tpr_knn, label="knn")
plt.plot(fpr_svm, tpr_svm, label="svc")
plt.plot(fpr_xgb, tpr_xgb, label="xgb")
plt.plot(fpr_cnn, tpr_cnn, label="cnn")
plt.plot(fpr_res, tpr_res, label="res")
plt.plot(fpr_vit, tpr_vit, label="vit")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.legend()
plt.title("ROC curves for all models on Step 1")
```
